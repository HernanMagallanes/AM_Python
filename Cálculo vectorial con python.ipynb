#### Producto punto entre vectores 

$\displaystyle \vec{A} \cdot \vec{B} = a_x\cdot b_x + a_y\cdot b_y +a_z\cdot b_z$

$\displaystyle \| \vec{A} \cdot \vec{B} \| = \| \vec{A} \|  \| \vec{B} \| cos(\theta)$

si $ \| \vec{A}\| = \| \vec{B} \| =1 $ el producto punto da una idea del ángulo entre vectores, variando entre $-1\leq cos(\theta)\leq 1$

si $cos(\theta)= 1$ los vectores tienen la misma dirección e igual sentido, en cambio si $cos(\theta)= -1$ los vectores tienen la misma dirección pero sentido contrario.

En el caso que $cos(\theta)= 0$ los vectores son perpendiculares (osea que no hay proyección)

```python
A = Matrix([[1,1,1]])
B = Matrix([[2,1,1]])

# A . B =
A.dot(B)
```

```python
# vectores normalizados

a = A.normalized()
b = B.normalized()

N(a.dot(b),3)
```

#### Producto cruz entre vectores

$\displaystyle \vec{A} \times \vec{B} = \begin{vmatrix}
\hat{i} & \hat{j} & \hat{k}\\
a_x & a_y & a_z \\
b_x & b_y & b_z
\end{vmatrix} $

Devuelve un vector perpendicular a los otros dos vectores. 

$\displaystyle \| \vec{A} \times \vec{B} \| = \| \vec{A} \|  \| \vec{B} \| sin(\theta)$

La norma de este nuevo vector representa el área del paralelogramo entre los dos vectores.

si $ \| \vec{A}\| = \| \vec{B} \| =1 $ entonces  el área varia entre $0\leq sin(\theta)\leq 1$

si $sin(\theta)= 0$ vectores con igual dirección y sentido, en cambio si $sin(\theta)= 1$ ambos vectores son perpendiculares, se representa el área de un cuadrado.

Los vectores no perpendiculares, el área disminuye (paralelogramo estirado)

```python
A = Matrix([[1,1,0]])
B = Matrix([[1,-1,0]])

# A X B =
A.cross(B)
```

```python
#|A X B| =
A.cross(B).norm()
```

#### Recta y plano

Si se tiene un punto $\vec{P}_o$ y un vector se pueden construir rectas y planos en el espacio con un parametro $t$.

el vector $\vec{V}$ le da la dirección a la recta (vector  director)

Para una recta: $ \vec{X} = \vec{P}_o + t \cdot \vec{V}$

Para un plano: $ \vec{X} = \vec{P}_o + t_1 \cdot \vec{V} + t_2 \cdot \vec{U}$

el vector $\vec{n}$ es perpendicular al plano (vector norma) tal que $\vec{V} \times \vec{U} = \vec{n} $


$ \vec{n} \cdot ( \vec{X} - \vec{P}_o )=0$

### 1. Curvas 

Si $\vec R : \mathbb{R} \rightarrow :\mathbb{R}^3 $ se define una curva en el espacio según el parámetro $t$ en un intervalo $[a,b]$

$\vec R(t) = f(t) \hat i + g(t) \hat j + h(t) \hat k $

Es una curva regular o suave cuando su función vectorial asociada tiene derivadas continuas y no nulas en el intervalo, una curva puede ser suave a trozos.

Las curvas se clasifican en cerradas cuando $\vec R(a) = \vec R(b)$ o abiertas cuando tiene un inicio y un final en diferentes puntos del espacio.

También se dice que un curva en simple si solo si es inyectiva en el intervalo (no se cruza sobre sí misma)

A las curvas que son cerradas y simples se les denomina "curvas de Jordan", estas funciones vectoriles se usan en varios teoremas.

Se puede escribir $D_t \vec {R(t)}$

### d. vectores tangente,normal y binormal

Sea una curva $C$ con ecuación vectorial $R(t)$, para un punto se pueden definir los siguiente vectores.

vector tangente: 

$\displaystyle \vec T(t) = \frac{ \vec { R'(t)}}{\| \vec {R'(t)\|}} $

vector normal:

$\displaystyle \vec N(t) = \frac{\vec { T'(t)}}{\| \vec {T'(t)\|}}$

vector binormal: 

$\displaystyle \vec B(t) = \frac {\vec T(t) \times \vec N(t)}{\|\vec T(t) \times \vec N(t)\|}  $

Los tres son vectores unitarios y ortogonales entre si.

```python
t = symbols('t')

R_vec = Matrix([[t, 10*t, t**2]])
R_vec
```

```python
# Vector tangente

T_vec = diff(R_vec,t) 

# Para el punto en t=4
# normalizamos el vector, reemplamos t=4 y con float podemos graficar

t_vec = T_vec.normalized().subs(t,4)
```

```python
# Vector Normal

N_vec = diff(R_vec,t,2) 

n_vec = N_vec.normalized().subs(t,4)
```

```python
# Vector Binormal

B_vec = T_vec.cross(N_vec)

b_vec = B_vec.normalized().subs(t,4)
```

### Representación física

En la rama de la física de la cinética, las trayectorias $\vec R$ se usan para representar la posición de un objeto sólido (para un sólido continuo, se habla de la posición del centro de masa del objeto)

Si la trayectoria está parametrizada $\vec R(t)$  y la variable $t$ representa el tiempo entonces ...

Si derivamos al trayectoria, tenemos la velocidad $V(t) = \vec {R'(t)}$

$ \displaystyle \vec {R'(t)} = \| R'(t) \| \cdot \vec {T(t)}  $

Si derivamos una vez más, tenemos la aceleración $A(t) = \vec {R''(t)}$

$ \displaystyle \vec {R''(t)} = \big[ \| R''(t) \| \big] \vec {T(t)} + \big[ \| R'(t) \| \big] \vec {T'(t)} $

Reemplazamos por el vector normal

$ \displaystyle \vec {R''(t)} = \big[ \| R''(t) \| \big] \vec {T(t)} + \big[ \| R'(t) \| \cdot \| T'(t) \|\big] \vec {N(t)} $

Tenemos que $\vec {R'}$ representa la magnitud de la velocidad del objeto en la dirección del vector tangente a la curva $\vec T$, como cambia la posición del objeto respecto al tiempo.

Por otro lado tenemos que $\vec {R''}$ representa la magnitud de la aceleración del objeto la cual tiene una componente tangencial y otra componente normal a la trayectoria, como cambia su velocidad respecto al tiempo.

$ A(t) = a_T \hat T + a_N \hat N$

Tanto las componentes, como la magnitud de cada una, son funciones del tiempo.

$L = \displaystyle \int_{a}^{b} ||R'(t)|| dt = \int_{a}^{b} \sqrt {[f'(t)]^2 + [g'(t)]^2 + [h'(t)]^2} dt$

### 2. Superficies

Una superficie se puede expresar en tres formas: explícita, implícita o parametrizada, siendo esta ultima una expresion vectorial. 

Forma explícita $S = \{ x,y,z= f(x,y) \}$ 

Forma implícita $ S = \{ (x,y,z) | h(x,y,z)=0 \}$

Forma paramétrica, sea una región del plano $\mathbb{R}^2$  se puede definir una ecuación vectorial  $\vec F: \mathbb{R}^2 \rightarrow \mathbb{R}^3$

Con parametrización $\displaystyle u,v \rightarrow \vec F = \left( x(u,v), y(u,v), z(u,v) \right)$ 

Se llama superficie simple si los vectores $\vec r_u$ y $\vec r_v$ son linealmente independientes, tal que

$ \displaystyle \vec r_u = \frac{\partial \vec F }{\partial u} = \left( \frac{\partial x}{\partial u} ,\frac{\partial y}{\partial u},\frac{\partial z}{\partial u}\right)$ 

$ \displaystyle \vec r_v = \frac{\partial \vec F }{\partial v} = \left( \frac{\partial x}{\partial v} ,\frac{\partial y}{\partial v},\frac{\partial z}{\partial v}\right)$

Para el caso que $u = x$ , $v=y$ con $z = f(x,y)$ queda:

$ \displaystyle \vec r_u = \frac{\partial \vec F }{\partial u} = \left(1,0,\frac{\partial z}{\partial u}\right)$ 

$ \displaystyle \vec r_v = \frac{\partial \vec F }{\partial v} = \left(0,1,\frac{\partial z}{\partial v}\right)$

### a. Área de una superficie

La forma parametrizada de las superficies se utiliza para calcular áreas no planas en el espacio.

Mientras que para superficie plana se puede aplicar una integral simple o doble pero para superficies no planas, se emplea el siguiente método.

Se puede teselar la superficie en trozos muy pequeños $dS$ y esta área se puede representar con la norma del vector normal.

$\displaystyle A = \iint_{S} \vec {dS} = \iint_{S} \| \vec{n} \| dS  $

Se nos presenta el diferencial de superficie $dS$ si bien podemos decir que es $dxdy$ hay casos en que las superficies no se proyectan en el plano $xy$.

Hay que tener en cuenta sobre qué plano coordenado se puede "proyectar" toda la superficie.

También se tiene que definir el vector normal a la superficie $\vec n$ con un producto cruz.

$\vec {n} = \vec r_u \times \vec r_v $

como versor normal

$\displaystyle \hat {n} = \frac{ \vec r_u \times \vec r_v}{ \| \vec r_u \times \vec r_v \|}  $


```python
# vamos a comenzar un ejemplo simple, áreas alabeadas en el espacio (planos) sobre el plano xz 
# voy a usar la forma implícita F(x,y,z)=0 para aprovechar el vector normal. 

def plano(t):
    y, z = np.meshgrid(t,t)
    x = 1 -y
    return x,y,z
```

### c. Orientación de una superficie

Se puede decir que una superficie es "orientable" si es posible decidir sin ambigüedad cual es cada uno de los lados.

La orientación se vee con el versor normal.
$\displaystyle \hat {n} = \frac{\nabla f(x,y,z)}{\|\nabla f(x,y,z)\|} $

Si $\hat {n}$ apunta para afuera se dice que la superficie está orientada positivamente, pero si consideramos la dirección opuesta, se dice que la superficie está orientada negativamente. Uno apunta hacia afuera, otro para adentro.

```python
from sympy import Function

x,y = symbols('x y')
f = Function('f')(x,y)
F = Matrix([[x, y, f]])
F
```

```python
# vector normal a la sup. de forma explícita z=f(x,y)

diff(F,x).cross(diff(F,y))

# tenemos +n o -n puede ser saliente o entrante
```

```python
diff(F,x).cross(diff(F,y)).norm() # sympy aplica el modulo en cada componente  
```

```python
# para graficar vamos a definir funciones para facilitar las cosas.

def sup_esf(r,p,t):
    
    ''' 0< p <2pi // 0< t <pi  '''
    
    x = r*np.cos(p)*np.sin(t)
    y = r*np.sin(p)*np.sin(t)
    z = r*np.cos(t)
    
    return x,y,z

def normal_esf(r,p,t):
    
    xn = (np.sin(t)**2)*np.cos(p)* r**2 
    yn = (np.sin(t)**2)*np.sin(p)* r**2
    zn = np.cos(t)*np.sin(t)* r**2
    
    return xn,yn,zn
```

Matriz hessiana asociada al campo escalar $f(x,y)$. 

$ \mathcal{H} f =  \large  \begin{pmatrix}{ \frac{\partial^2{f}}{\partial{x^2}}  }&{ \frac{\partial^2{f}}{\partial{x} \partial{y}} }\\{ \frac{\partial^2{f}}{\partial{y} \partial{x}} }&{ \frac{\partial^2{f}}{\partial{y^2}} }\end{pmatrix} $

Esta matriz para cualquier campo escalar debe ser simétrica, por teorema de Schwarz (igualdad de las derivadas cruzadas)

La igualdad de derivadas cruzadas es la condición de simetría que nos sirve como un condición de integrabilidad.

Entonces si tenemos un $\vec F(x,y)$ podemos encontrar un campo escalar $f(x,y)$ asociado si se cumple esta condición.

$\displaystyle \frac{\partial^2{f}}{\partial{x} \partial{y}} = \displaystyle \frac{\partial^2{f}}{\partial{y} \partial{x}} $ 

Si tenemos una función vectorial $\vec{F} = P(x,y) \hat i + Q(x,y) \hat j  $

con $P = f(x,y)$ , $ Q = g(x,y)$ funciones continuas en un disco de $\mathbb{R}^2$ (plano xy) y con derivadas continuas.

$\vec F$ es un gradiente de un campo escalar $f(x,y)$ si solo si

$ \displaystyle \frac{P(x,y)}{\partial{y}} = \frac{Q(x,y)}{\partial{x}}$


se puede encontrar un función:

$ \displaystyle f(x,y) = \int P dx + g(y) $

$ \displaystyle f(x,y) = \int Q dx + h(x) $

Esta condición es propiedad de los campos $\vec F$ conservativos. 

Los campos escalares $f(x,y)$ resultantes son funciones potenciales, generalmente se denotan $\phi(x,y)$

$\phi(x,y) = f(x,y) + g(y) + C$

## B. Operadores vectoriales

Ya se vio el operador gradiente $\nabla$ de un campo escalar $f(x,y)$

$ \displaystyle  \vec \nabla f = \frac{\partial}{\partial x} f(x,y) \hat i + \frac{\partial}{\partial y} f(x,y) \hat j $


Para un campo $\vec F$ se aplica el gradiente para cada componente, formando un matriz, llamada matriz jacobiana 

Si $\vec F = F_x(x,y,z)\hat i + F_y(x,y,z) \hat j + F_z(x,y,z)  \hat k $

$\displaystyle \nabla \vec F = D \vec F $

$ \displaystyle   D \vec F = \begin{bmatrix}
\nabla F_x \\ \nabla F_y \\ \nabla F_z \\
\end{bmatrix}  = \begin{bmatrix}
\frac{\partial}{\partial x}F_x & \frac{\partial}{\partial y}F_x & \frac{\partial}{\partial z}F_x \\
\frac{\partial}{\partial x}F_y & \frac{\partial}{\partial y}F_y & \frac{\partial}{\partial z}F_y \\
\frac{\partial}{\partial x}F_z & \frac{\partial}{\partial y}F_z & \frac{\partial}{\partial z}F_z \\
\end{bmatrix}$

mide la tasa y dirección del cambio de un campo, en que direccion aumenta cada componente.

```python
x,y,z = symbols('x y z')
F = Matrix([ x*y, y*z , z*x ])
F
```

```python
# gradiente de un campo vectorial (jacobiano)

X = Matrix([x, y, z])
J = F.jacobian(X)
J
```

```python
# determinante del jacobiano (cuando la matriz es cuadrada)

J.det()
```

### 1. Rotor o rotacional de un campo vectorial

Si tenemos un campo $\vec F (x,y,z)$ se puede definir su rotacional


$\displaystyle rot (\vec F) = \vec \nabla \times \vec F = \left( \frac{\partial}{\partial x} , \frac{\partial}{\partial y},\frac{\partial}{\partial z} \right) \times \left( F_x,F_y,F_z \right) = \begin{vmatrix}
\hat{i} & \hat{j} & \hat{k}\\
\frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \\
F_x & F_y & F_z
\end{vmatrix} $

Es una magnitud vectorial que representa la tendencia de un campo vectorial en inducir rotación, cuanto "gira" en un punto.

Al aplicar este operador a un campo vectorial $\vec F_1 (x,y,z)$, se obtiene un campo vectorial nuevo $\vec F_2 (x,y,z)$

Si se tiene un campo escalar $f(x,y)$ y tiene segundas derivadas continuas, entonces $rot( \vec \nabla f ) = 0$

```python
x,y,z = symbols('x y z')

F = Matrix([ E**(2*x), 3*(x**2)*y*z , x+ (2*y**2)*z ])
F
```

```python
# Rotor / rotacional

rx = diff(F[2],y) - diff(F[1],z)

ry = diff(F[0],z) - diff(F[2],x)

rz = diff(F[1],x) - diff(F[0],y)

Rot = Matrix([rx, ry, rz])
Rot
```

### 2. Divergencia de un campo vectorial

Si tenemos un campo $\vec F (x,y,z)$ se puede definir su divergencia  


$\displaystyle div (\vec F) = \vec \nabla \cdot \vec F  = \left( \frac{\partial}{\partial x} , \frac{\partial}{\partial y},\frac{\partial}{\partial z} \right) \cdot \left( F_x,F_y,F_z \right) = \frac{\partial F_x}{\partial x}  + \frac{\partial F_y}{\partial y} + \frac{\partial F_z}{\partial z} $


Es una magnitud escalar que representa sumideros o fuentes en el espacio, el signo negativo es un sumidero donde el campo desaparece/destruye, por el contrario, si en un punto tiene signo positivo es una fuentes donde el campo aparece/crea.

Al aplicar este operador sobre $\vec F(x,y,z)$, se obtiene un campo escalar $f(x,y,z)$

Una propiedad a destacar: $div \left( rot(\vec F) \right) =0$

Los campos vectoriales con divergencia nula $div =0$ en todo sus puntos se denominan "solenoidal".

Ejemplos en ramas distintas de física, campos magnéticos y campos de velocidades de fluidos incompresibles.

```python
x,y,z = symbols('x y z')

F = Matrix([ E**(2*x), 3*(x**2)*y*z , x+ (2*y**2)*z ])
F
```

```python
# divergencia del campo

div = diff(F[0],x) + diff(F[1],y) +diff(F[2],z)
div
```

### 3. Laplaciano

El operador divergencia se puede extender al operador laplaciano:

$\displaystyle \nabla \cdot\nabla f= \nabla ^2 f = \frac{\partial ^2 F_x}{\partial^2 x}  + \frac{\partial^2 F_y}{\partial^2 y} + \frac{\partial^2 F_z}{\partial^2 z}$

Si una función tiene un laplaciano nulo en todo el espacio, es una función armónica
$\displaystyle \nabla ^2 f = 0$


Propiedad para campos vectoriales:

$\displaystyle \displaystyle \nabla ^2 \vec F = \vec \nabla div\left( \vec F\right) - rot\left( rot(\vec F) \right)    $

### 1. Integral de linea

Es la interacción de un campo vectorial $\vec F$ con una curva $C$ representada con una trayectoria $R(t)$

$ \displaystyle \int_C \vec {F} \cdot \vec{dr}  =  \int_{a}^{b} \vec {F( R(t))} \cdot  {\vec R'(t)}    dt$

La definición de integral de linea se aplica para una curva suave o para una curva suave a trozos.

Representa el trabajo realizado por un campo vectorial (campo de fuerza) al desplazar una partícula que recorre dicha trayectoria, si es positivo el campo entrega trabajo, en cambio si es negativo hay que aportar trabajo para realizar el camino, el campo es desfavorable.

Si la trayectoria describe una curva cerrada, se habla de circulación, en estos casos se tiene que tener en cuenta la orientación positiva de la curva.

$\displaystyle \oint_{C^+} \vec F \cdot \vec{dr} = \int_{C_1} \vec F \cdot \vec{dr} + \int_{C_2} \vec F \cdot \vec{dr}  $

```python
# Ejemplo 1 trabajo de una campo F = x i sobre la trayectoria recta r = x i  (linea paralela  al eje x desde x=0 a x=1) 

# trayectoria parametrizada - recta paralela al eje x
t = symbols('t')
R = Matrix([[ t,0,0 ]])

R_t = diff(R,t)

# campo vectorial - con única componente i
x,y,z = symbols('x y z')
F = Matrix([ x,0,0 ])
F
```

```python
# parametrizamos el campo con la trayectoria

F_R_t = F.subs(x,R[0]).subs(y,R[1]).subs(z,R[2])
F_R_t
```

```python
# producto punto entre el campo parametrizado F(R(t)) y el vector tangente R'(t)

FR = F_R_t.dot(R_t)

# integral desde a=0 a b=1

Integral_linea = Integral(FR,(t,0,1))
Integral_linea
```

$\displaystyle \int_{0}^{1}t dt $ 

```python
# trabajo realizado por el campo sobre esa trayectoria
# para este primer ejemplo el campo y la trayectoria son paralelos en todos los puntos 

Integral_linea.doit()
```

Si el campo vectorial es conservativo, tiene una función potencial asociada $\phi(x,y)$ con siguiente propiedad:


$\displaystyle \int_{a}^{b} \vec F \cdot \vec{dr} = \phi(b) - \phi(a) $

Significa que no importa la trayectoria, hay independencia del camino y se calcula como una diferencia de valores de la función potencial asociada.

Para el caso de que la curva $C$ sea cerrada $\displaystyle \rightarrow \oint_{C} \vec{F}\cdot\vec{dr} =0 $ 

El concepto de función potencial e integral de linea tiene aplicaciones en varias ramas de la física como en energía mecánica, termodinámica y electricidad.

Se puede ver mejor la relación entre $\vec F$ y $\phi$ en este caso, el trabajo de $\vec F$ se calcula como diferencia entre potenciales $\phi$.

En física, en algunos casos, se cambian el "punto de vista" y se usa la siguiente forma:   $\displaystyle \vec{F}=- \vec\nabla \phi $

Como en los potenciales eléctricos, o son proporcionales como en los procesos de difusión de una sustancia y en el flujo de calor.

### 2. Integral de flujo

Es la interacción de un campo vectorial $\vec F $ con una superficie $S$ con vector normal $\hat{n}$ es saliente a la superficie (cerrada o abierta)

$\displaystyle \iint_{S} \vec F \cdot \vec{dS} = \iint_S \vec F \cdot \hat{n} dS = \iint_S \vec F \left( \vec r (u,v)\right) \cdot \left( \vec r_u \times \vec r_v \right) dudv $

Representa la cantidad de campo vectorial que atraviesan el área.

Como se calcula con un producto punto entre el campo $\vec F$ y el versor normal $\hat n$ de la superficie, el flujo es una magnitud escalar. 

```python
# Ejemplo flujo del campo F sobre una porción del plano y = 1

# superfice - plano
x, y = symbols('x y')
sup = Matrix([x, y, 1])

# vector normal
n = Matrix([0,1,0])

# campo vectorial
x,y,z = symbols('x y z')

F = Matrix([0,1,0])

# producto escalar entre el campo y la normal
f = F.dot(n)

I1 = Integral(f,(z,-1,1))
I2 = Integral(I1,(x,-1,1))  
I2
```

### 3. Definiciones de operadores vectoriales
#### Definicion del operador rotor

Para un campo $\vec F (x,y,z)$ se puede definir su rotacional:

$\displaystyle rot( \vec F) = \vec \nabla \times \vec F = \lim \limits_{S \to 0} \frac{1}{S} \oint_{\partial S} \vec F \cdot \vec {dr} $

El operador rotor es el límite del cociente entre la integral de línea de $\vec F$ sobre una curva cerrada $\partial S$ osea la frontera de una superficie $S$ y el área de dicha superficie, cuando el área de esta tiende a cero.

En esta definición solo calcularía una componente del rotacional, la componente paralela al vector normal $\vec n$ de la superficie $S$ , para el rotacional completo se deben tomar tres componentes ortogonales.

#### Definicion del operador divergecia

Para un campo $\vec F (x,y,z)$ se puede definir su divergencia :

$\displaystyle div ( \vec F) = \vec \nabla \cdot \vec F  = \lim \limits_{V \to 0}\frac{1}{V} \iint_{S} \vec F \cdot \vec {dS} $

El operador divergencia es el límite del cociente entre una integral de flujo de $\vec F$ sobre una superficie cerrada $S$ y el volumen que encierra dicha superficie, cuando este volumen tiene a cero.

Hay que resaltar que se utiliza el vector normal $\vec n$ saliente a la superficie.

## D. Teoremas integrales

Sirven para facilitar el cálculo de integrales, el teorema de Green y de Stokes relaciona las integrales de línea con integrales dobles, mientras que el teorema de Gauss relaciona las integrales de flujo con integrales triples. 

### 1. Teorema de  Green 

Si tenemos un campo $\vec F = P(x,y) \hat i + Q(x,y) \hat j $ si son funciones con derivadas parciales continuas en el plano y una curva $C$ cerrada, simple y suave o suave a trozos.

La integral de línea de $\vec F$ sobre la curva $C$ orientada positivamente, se puede calcular como una integral doble sobre la región $D$ delimitada por la curva, de la siguiente forma:

$\displaystyle \oint_{C^+} \vec {F} \cdot \vec{dr} = \iint_D \left( \frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} \right)  dx dy$

Generalmente se usa la siguiente forma para explicitar las componentes del campo:

$\displaystyle \oint_{C^+} P(x,y) dx + Q(x,y) dy = \iint_D \left( \frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} \right)  dA$

Nos ahorramos la parametrización de cada trozo de la curva en el campo.

```python
## Ejemplo 1 campo F sobre la curva cerrada C : x=cos(t), y=sin(t) (circunferencia de radio unitario)

## integral de línea
####################

# curva parametrizada
t = symbols('t')

R = Matrix([[cos(t), sin(t)]])
R_t = diff(R,t)

# campo vectorial F=(P,Q)
x,y = symbols('x y')
F = Matrix([-y, x])

# incluimos la trayectoria en el campo
F_R_t = F.subs(x, R[0]).subs(y, R[1])

# producto entre F(R(t)) .R'(t)
FR = F_R_t.dot(R_t)

# integral desde 0 a 2 pi 

Int_linea = Integral(FR,(t,0, 2*pi))
Int_linea
```

### 2. Teorema de  Stokes

Si tenemos un campo $\vec F$ con derivadas parciales continuas y una superficie $S$ suave o suave a trozos con vector normal saliente $\hat n$ y que su frontera se pueda definir una curva cerrada $\partial S$ positivamente.

La integral de línea de $\vec F$ sobre la curva $\partial S$ orientada positivamente, se puede calcular como una integral de flujo del rotacional de $\vec F$ sobre la superficie $S$. 

$\displaystyle \oint_{\partial S^+} \vec F \cdot \vec dr  =  \iint_S rot(\vec F) \cdot dS $

La ventaja del teorema es que nos ahorramos una integral de circulacion por cada arista de la superficie para las cuales tendriamos que definir cada frontera como una curva y parametrizar el campo con esta nueva curva, nos evitamos todo este trabaja con una integral doble.

Si la curva $\partial S$ esta orientada positivamente, el vector normal apunta para adentro y el vector binormal es paralelo a la normal de la superficie. 

```python
# básicamente el teorema de Stokes es el teorema de Green en tres dimensiones.
```

### 3. Teorema de  Gauss 

De nuevo, para un campo $\vec F$ con derivadas parciales continuas y una superficie $S$ cerrada y suave o suave a trozos que encierra un volumen,  con vector normal $\vec n$ saliente a la superficie.

La integral de flujo de $\vec F$ sobre la superficie $S$ se puede calcular como una integral triple de la divergencia del campo, usando el volumen encerrado.

$\displaystyle \iint_S \vec F \cdot \vec {dS} = \iiint_V div(\vec F) dV$

La ventaja del teorema está en el caso de las superficies compuestas de n trozos tendríamos que calcular n integrales de flujo (por ejemplo, para tetraedro tendriamos 4 integrales de flujo o un cubo 6 integrales) nos las ahorramos con una sola integral triple con el sólido a los límites.

```python
# la integral doble es CERRADA, por lo general aparece con un círculo en el símbolo de integral.
```

```python
# Ejemplo 1 flujo del campo F sobre la superficie de la esfera de radio =4 

# Integral de flujo
###################

# campo vectorial
x,y,z = symbols('x y z')
F = Matrix([0,0, 5*z])

# esfera - parte positiva
sup = Matrix([x, y, sqrt( 16 -x**2 -y**2)])


# vector normal
f_u = diff(sup,x)
f_v = diff(sup,y)
n = f_u.cross(f_v)

f_r = F.subs(x,sup[0]).subs(y,sup[1]).subs(z,sup[2])
f = f_r.dot(n)

# coordenadas polares
r, t = symbols('r theta')
f_pol = f.subs(x,r*cos(t)).subs(y,r*sin(t))


I1 = Integral(f_pol.simplify()*r ,(r,0,4)) # jacobiano por polares 
I2 = Integral(I1,(t,0,2*pi))  
I2
```

$\displaystyle \int_{0}^{2 \pi} \int_{0}^{4} 5 r \sqrt{16 - r^2} dr d\theta$ 

```python
a = I2.doit()
a
```
